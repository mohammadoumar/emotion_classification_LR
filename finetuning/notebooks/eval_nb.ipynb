```python
grounds = [json.loads(x)["list_emotion_classes"] for x in grounds]
```

```python
grounds = [['Neutral'] if elem == [['Neutral']] else elem for elem in grounds]
predictions = [['Neutral'] if elem == [['Neutral']] else elem for elem in predictions]
```

```python
len(grounds), len(predictions)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()
```

```python
y_true_mhot = mlb.fit_transform(grounds)
```

```python
y_true_mhot.shape
```

```python
y_pred_mhot = mlb.transform(predictions)
```

```python
y_pred_mhot.shape
```

```python
class_labels = mlb.classes_
```

```python
class_labels
```

```python
print(classification_report(y_true_mhot, y_pred_mhot, target_names=class_labels, digits=3))
```

```python
with open("/Utilisateurs/umushtaq/emotion_analysis_comics/finetuning/saved_models/comics35_Meta-Llama-3.1-8B-Instruct-bnb-4bit//classification_report_2.pickle", 'wb') as fh:
    
    pickle.dump(classification_report(y_true_mhot, y_pred_mhot, target_names=class_labels, digits=3, output_dict=True), fh)
```
