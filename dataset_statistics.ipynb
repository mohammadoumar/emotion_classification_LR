```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("/Utilisateurs/umushtaq/emotion_analysis_comics/dataset_files/comics_dataset_w_spanish.csv")
```

```python
def get_emotions(row):

    utterance_emotions = row.emotion
    utterance_emotions_l = []
    emotion_class_labels = ["Anger", "Disgust", "Fear", "Sadness", "Surprise", "Joy"]

    if utterance_emotions == 'Neutral':
        
        utterance_emotions_l.append(utterance_emotions)
    
    else:
        utterance_emotions = utterance_emotions.split("-")

        for idx, emotion_annotation in enumerate(utterance_emotions):

            if '0' not in emotion_annotation:
                
                utterance_emotions_l.append(emotion_class_labels[idx])                

    return utterance_emotions_l
```

```python
df['emotion_u'] = df.apply(lambda row: get_emotions(row), axis=1)
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Download tokenizer if not already available
nltk.download("punkt_tab")

# Ensure the column exists and contains strings
if "utterance" in df.columns:
    df["utterance"] = df["utterance"].astype(str)  # Convert to string to handle NaN
    
    # Tokenize each utterance into words and count tokens
    df["utterance_word_count"] = df["utterance"].apply(lambda x: len(word_tokenize(x)))

    # Compute the average word count
    average_word_length = df["utterance_word_count"].mean()
    media_word_length = df["utterance_word_count"].median
```

```python
if "emotion_u" in df.columns:
    # Compute the length of each list
    df["emotion_u_length"] = df["emotion_u"].apply(len)
    
    # Count occurrences of each unique length
    length_counts = df["emotion_u_length"].value_counts().sort_index()
```

```python
if "emotion_u" in df.columns:
    # Flatten all lists into a single list
    all_emotions = [emotion for sublist in df["emotion_u"] for emotion in sublist]

    # Count occurrences of each unique emotion
    emotion_counts = pd.Series(Counter(all_emotions)).sort_values(ascending=False)
```
