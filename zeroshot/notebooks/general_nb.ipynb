````python
import json
from sklearn.preprocessing import MultiLabelBinarizer

def get_mlb(grounds, predictions):
    
    mlb = MultiLabelBinarizer()
    grounds_mhot = mlb.fit_transform(grounds)
    predictions_mhot = mlb.transform(predictions)
    
    return grounds_mhot, predictions_mhot, mlb.classes_
````

````python
grounds_mhot, predictions_mhot, classes = get_mlb(grounds, predictions_l)
````

````python
print(classification_report(grounds_mhot, predictions_mhot, target_names=classes, digits=3))
````

````python
from pathlib import Path
````

````python
classification_file = Path("/Utilisateurs/umushtaq/emotion_analysis_comics/zeroshot/results/comics35_zs_Phi-3-mini-4k-instruct-bnb-4bit/") / "classification_report.pickle"

with classification_file.open('wb') as fh:
    
    pickle.dump(classification_report(grounds_mhot, predictions_mhot, target_names=classes, output_dict=True), fh)
````
