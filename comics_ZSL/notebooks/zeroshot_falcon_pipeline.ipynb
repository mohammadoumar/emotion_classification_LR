```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch
from tqdm.notebook import tqdm
```

```python
model = "tiiuae/falcon-7b-instruct"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    # trust_remote_code=True,
    device_map="auto",
)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/Utilisateurs/umushtaq/emotion_analysis_comics/zeroshot/datasets/comics_data_processed.csv")
df = df.drop(columns=[df.columns[0], df.columns[1]])
```

```python
def make_prompt(text):

    formatted_string = (
        f"You are an expert sentiment analysis assistant that takes an utterance from a comic book "
        f"and must classify the utterance into the following emotion labels(es): anger, surprise, fear, disgust, sadness, joy, neutral."
        f"Utterance:{text}"
        f'Your answer must be in the following JSON format: '
        f'{{"utterance_emotions": emotion_class}}. For example: '
        f'{{"utterance_emotions": "anger"}}'
        f"It is very important that you only return the JSON object and nothing else!\n"
        f"Answer: "
        #f"Here is the JSON: ____ where the blank is replaced by "
    )

    return formatted_string
```

```python
texts = df.utterance.tolist()
texts = [make_prompt(text) for text in texts]
```

```python
len(texts)
```

```python
sequences = []

for i, text in tqdm(enumerate(texts)):
    print(i)
    sequence = pipeline(
    #"Girafatron is obsessed with giraffes, the most glorious animal on the face of this Earth. Giraftron believes all other animals are irrelevant when compared to the glorious majesty of the giraffe.\nDaniel: Hello, Girafatron!\nGirafatron:",
        text,
        #ax_length=200,
        do_sample=True,
        #top_k=10,
        num_return_sequences=1,
        max_new_tokens=32,
        eos_token_id=tokenizer.eos_token_id,
    )
    sequences.append(sequence)
```

```python
cleaned_seqs = []

for sequence in sequences:
    cleaned_seqs.append(sequence[0]['generated_text'].split("Answer:")[1])
```
