```python
## Now get p value
```

```python
def paired_bootstrap_weighted_f1(y_true, y_zs, y_ft, B=10000):
    # Convert to numpy arrays (CRITICAL FIX)
    y_true = np.asarray(y_true)
    y_zs   = np.asarray(y_zs)
    y_ft   = np.asarray(y_ft)

    n = len(y_true)
    deltas = np.empty(B)

    for b in range(B):
        idx = np.random.randint(0, n, n)

        f1_zs = f1_score(y_true[idx], y_zs[idx], average="weighted")
        f1_ft = f1_score(y_true[idx], y_ft[idx], average="weighted")

        deltas[b] = f1_ft - f1_zs

    mean_delta = deltas.mean()
    ci_low, ci_high = np.percentile(deltas, [2.5, 97.5])
    p_value = 2 * min(np.mean(deltas <= 0), np.mean(deltas >= 0))

    return mean_delta, (ci_low, ci_high), p_value
```

```python
mean_delta, (ci_low, ci_high), p_value = paired_bootstrap_weighted_f1(y_true, y_zs, preds_ft, B=10000)
```
